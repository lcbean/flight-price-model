```python
from scipy import stats


# calculate z-scores to plot outliers
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))

outlier_condition = (z_scores > 3).any(axis=1) | (z_scores < -3).any(axis=1)
df_outliers = df[outlier_condition]

# make box plots for each column 

for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.pyplot.figure(figsize=None)
    sns.boxplot(y=df[column], palette='crest')
    plt.pyplot.title(f'{column}', fontsize=14)
    plt.pyplot.ylabel(column, fontsize=12)
    plt.pyplot.show()
```

```python
# remove outliers
df = df[~outlier_condition]
```

# Feature Scaling

```python
x = df.drop(['fare'], axis=1)
y = df['fare']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data1 = scaler.fit_transform(x)

data1[0]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=123)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## ExtraTreesRegressor Model

```python
selection = ExtraTreesRegressor()
selection.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def metrics(y_act, y_pred):
    print(f'RMSE: ', mean_squared_error(y_act, y_pred) ** 0.5)
    print(f'R_Squared_Value: ', r2_score(y_act, y_pred))

def accuracy(y_act, y_pred):
    errors = abs(y_act - y_pred)
    mape = 100 * np.mean(errors / y_act)
    accuracy = 100 - mape
    return accuracy
```

```python
y_pred = selection.predict(x_test)
```

```python
metrics(y_test, y_pred)
```

```python
accuracy(y_test, y_pred)
```

```python


plt.pyplot.scatter(y_test, y_pred)
plt.pyplot.xlabel('Actual Flight Fare')
plt.pyplot.ylabel('Predicted Flight Fare')
plt.pyplot.title('Prediction VS Actual Fare')
plt.pyplot.show()
```
